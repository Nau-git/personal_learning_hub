# Linear Regression

popular cost function in linear regression --> `MSE = (1/n) * Σ(y_i - ŷ_i)²` 

Where:

- `n` is the number of data points
- `y_i` is the actual value
- `ŷ_i` is the predicted value
- `Σ` represents the sum over all data points

```python
def mean_squared_error(y_true, y_pred):
    """
    Calculate the Mean Squared Error (MSE) between true and predicted values.
    
    Args:
    y_true (list): List of true values
    y_pred (list): List of predicted values
    
    Returns:
    float: The Mean Squared Error
    """
    if len(y_true) != len(y_pred):
        raise ValueError("The length of y_true and y_pred must be the same.")
    
    n = len(y_true)
    squared_errors = [(y_true[i] - y_pred[i])**2 for i in range(n)]
    mse = sum(squared_errors) / n
    return mse
```

```python
y_true = [2, 4, 5, 4, 5]
y_pred = [1.8, 3.9, 5.1, 4.2, 5.3]

mse = mean_squared_error(y_true, y_pred)

print(f"Mean Squared Error: {mse:.4f}")
```

```python
import matplotlib.pyplot as plt
def visualize_mse(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    squared_errors = [(y_true[i] - y_pred[i])**2 for i in range(len(y_true))]
    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(5, 7))
    
    # Plot true vs predicted values
    ax1.plot(range(len(y_true)), y_true, 'bo-', label='True values')
    ax1.plot(range(len(y_pred)), y_pred, 'ro-', label='Predicted values')
    for i in range(len(y_true)):
        ax1.plot([i, i], [y_true[i], y_pred[i]], 'k--', alpha=0.3)
    ax1.set_title('True vs Predicted Values')
    ax1.set_xlabel('Data point')
    ax1.set_ylabel('Value')
    ax1.legend()
    
    # Plot squared errors
    ax2.bar(range(len(squared_errors)), squared_errors)
    ax2.set_title(f'Squared Errors (MSE: {mse:.4f})')
    ax2.set_xlabel('Data point')
    ax2.set_ylabel('Squared Error')
    
    plt.tight_layout()
    plt.show()
```

```python
visualize_mse(y_true, y_pred)
```
